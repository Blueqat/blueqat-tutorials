## Single Arbitrary Rotation Gate
The Arbitrary Rotation Gate can be rotated at any angle you like.

### Rx, Ry, Rz

Rx, Ry, Rz gates can be implemented by the following.

| Rx | Ry | Rz |
|:-:|:-:|:-:|
|`rx(θ)`|`ry(θ)`|`rz(θ)`|

$\theta$ is the rotation angle for each axis.

### Rx gate
$$
Rx(\theta) =
\begin{bmatrix}
\cos(\frac{\theta}{2}) & -i\sin(\frac{\theta}{2}) \\
-i\sin(\frac{\theta}{2}) & \cos(\frac{\theta}{2})
\end{bmatrix}
$$

```python
Circuit().rx(math.pi/2)[0].m[:].run(shots=100)
```

```python
Circuit().rx(math.pi/2)[0].run(backend="draw")
```

### Ry gate
$$
Ry(\theta) =
\begin{bmatrix}
\cos(\frac{\theta}{2}) & -\sin(\frac{\theta}{2}) \\
\sin(\frac{\theta}{2}) & \cos(\frac{\theta}{2})
\end{bmatrix}
$$

```python
Circuit().ry(math.pi/2)[0].m[:].run(shots=100)
```

```python
Circuit().ry(math.pi/2)[0].run(backend="draw")
```

### Rz gate
$$
Rz(\theta) =
\begin{bmatrix}
e^{-i\frac{\theta}{2}} & 0 \\
0 & e^{i\frac{\theta}{2}}
\end{bmatrix}
$$

```python
Circuit().rz(math.pi/2)[0].m[:].run(shots=100)
```

```python
Circuit().rz(math.pi/2)[0].run(backend="draw")
```

### U gate
U gate can be implemented by the following.


| U |
|:-:|
|`u(θ,φ,λ,γ)`|

## Change from previous U1, U2, U3 gates
If your exising code using older version of Blueqat, update the existing code as follows,

- U1(θ) → RZ(θ)
- U2(φ, λ) → U(π/2, φ, λ, -0.5 * (φ+λ))
- U3(θ, φ, λ) → U(θ, φ, λ, -0.5 * (φ+λ))

U gate has four parameters, so it can create an arbitrary state with global phase.

$$
U(\theta, \phi, \lambda, \gamma) = e^{i\gamma}  \begin{bmatrix}
\cos(\frac{\theta}{2}) & -e^{i\lambda} \sin(\frac{\theta}{2}) \\
e^{i\phi} \sin(\frac{\theta}{2}) & e^{i(\phi+\lambda)} \cos(\frac{\theta}{2}) 
\end{bmatrix}
$$

```python
Circuit().u(math.pi/2, 0, math.pi, -math.pi/2)[0].m[:].run(shots=100)
```

### Rxx, Ryy, Rzz
Rxx,Ryy,Rzz gates can be implemented by the following.

| Rxx | Ryy | Rzz |
|:-:|:-:|:-:|
|`rxx(θ)`|`ryy(θ)`|`rzz(θ)`|

### RXX gate
$$
Rxx(\theta) = \left(\begin{array}{cccc}
\cos\theta &0&0&-i\sin\theta\\
0& \cos\theta &-i\sin\theta&0\\
0&-i\sin\theta& \cos\theta &0\\
-i\sin\theta&0&0& \cos\theta
\end{array}\right)\\
$$

```python
Circuit(2).rxx(math.pi/2)[0,1].m[:].run(shots=100)
```

```python
Circuit(2).rxx(math.pi/2)[0,1].run(backend="draw")
```

### RYY gate
$$
Ryy(\theta) = \left(\begin{array}{cccc}
\cos\theta &0&0&i\sin\theta\\
0& \cos\theta &-i\sin\theta&0\\
0&-i\sin\theta& \cos\theta &0\\
i\sin\theta&0&0& \cos\theta
\end{array}\right)
$$


```python
Circuit(2).ryy(math.pi/2)[0,1].m[:].run(shots=100)
```

```python
Circuit(2).ryy(math.pi/2)[0,1].run(backend="draw")
```

### RZZ gate
$$
Rzz(\theta) = \left(\begin{array}{cccc}
e^{\frac{\theta}{2}i} &0&0&0\\
0& e^{-\frac{\theta}{2}i} &0&0\\
0&0& e^{-\frac{\theta}{2}i} &0\\
0&0&0& e^{\frac{\theta}{2}i}
\end{array}\right)
$$


```python
Circuit().rzz(math.pi/2)[0,1].m[:].run(shots=100)
```

```python
Circuit(2).rzz(math.pi/2)[0,1].run(backend="draw")
```
